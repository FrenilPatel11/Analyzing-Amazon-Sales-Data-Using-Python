```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df.info()
```

```python
df.isnull().sum()
```

# drop null values


```python
df.dropna(inplace=True)
```

```python
df['ship-postal-code']=df['ship-postal-code'].astype('int')
```

### Exploratory Data Analysis

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Size')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From above Graph you can see that most of the people buys M-Size
```

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=True)
```

```python
bar_Qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.barplot(x='Size',y='Qty', data=bar_Qty)
```

```python
#From above Graph you can see that most of the Qty buys M-Size in the sales
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Courier Status',hue= 'Status')
plt.show()
```

```python
#From above Graph the majority of the orders are shipped through the courier.
```

```python
df['Size'].hist() 
plt.show()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=20, edgecolor='red')
plt.show()
```

```python
#From above Graph you can see that most of the buyers are T-shirt
```

```python
B2B_db = df['B2B'].value_counts()
print(B2B_db)
```

```python
plt.pie(B2B_db, labels=B2B_db.index, autopct='%1.1f%%')
plt.show()
```

```python
# From above chart we can see that maximum i.e. 99.2% of buyers are retailers and 0.8% are B2B buyers
```

```python
a1 = df['Fulfilment'].value_counts()
print(a1)
```

```python
x_data = df['Category']  
y_data = df['Size'] 
plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_state = df['ship-state'].value_counts().head(10)
print(top_10_state)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

```python
#From above Graph you can see that most of the buyers are Maharashtra state
```

## The data analysis reveals that the business has a significant customer base in Maharashtra state, mainly serves retailers, fulfills orders through Amazon, experiences high demand for T-shirts, and sees M-Size as the preferred choice among buyers.
